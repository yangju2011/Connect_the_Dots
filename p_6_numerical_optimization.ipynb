```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from pylab import rcParams
rcParams['figure.figsize'] = 5,5
```

```python
def h_x2(x1, c, p):
    # x1**2 + x2**2 = c
    x2_sq = (c - abs(x1)**(float(p)))** (1/float(p))
    return x2_sq, -x2_sq
```

```python
c = 4
x1_list = np.linspace(-2,2,100)
x2_list = np.array([h_x2(x1, 4, 2) for x1 in x1_list])
```

```python
def plot_contour(x1_list, x2_list, lim, formula):
    plt.plot(x1_list, x2_list[:,0],c='r')
    plt.plot(x1_list, x2_list[:,1],c='r')
    plt.xlim([-lim,lim])
    plt.ylim([-lim,lim])
    plt.xlabel(r"$x_1$")
    plt.ylabel(r"$x_2$")
    plt.plot([0,0],[-lim,lim],c='black')
    plt.plot([-lim,lim],[0,0],c='black')
    plt.text(0,2.2,formula)#r"$x_1^2 + x_2^2 = 4$")
```

```python
plot_contour(x1_list, x2_list, lim=3, formula=r"$S_1:|x_1|^2 + |x_2|^2 = 4$")
a_x1 = 0.2
a_x2 = h_x2(a_x1, 4, 2)[0]
b_x1 = 1.8
b_x2 = h_x2(b_x1, 4, 2)[0]
plt.plot([a_x1,b_x1],[a_x2,b_x2],c='blue')
```

```python
x3_list = np.array([h_x2(x1, 2, 1) for x1 in x1_list])
```

```python
plot_contour(x1_list, x3_list, lim = 3, formula = r"$|x_1|+ |x_2| = 2$")
```

```python
x4_list = np.array([h_x2(x1, 1.5, 0.5) for x1 in x1_list])
```

```python
plot_contour(x1_list, x4_list, lim = 3, formula = r"$S_2:|x_1|^{0.5} + |x_2|^{0.5} = 1.5 $")
a_x1 = 0.2
a_x2 = h_x2(a_x1, 1.5, 0.5)[0]
b_x1 = 1.5
b_x2 = h_x2(b_x1, 1.5, 0.5)[0]
plt.plot([a_x1,b_x1],[a_x2,b_x2],c='blue')
```
